```python
def tower(N,from_tower,to_tower,alt_tower):
    if N==0:
        return
    tower(N-1,from_tower,alt_tower,to_tower)
    print("Move disk %d from tower %d to tower %d"
              %(N,from_tower,to_tower))
    tower(N-1,alt_tower,to_tower,from_tower)
tower(3,1,2,3)
```

快速排序

```python
def quick_sort(lst):
    if len(lst)<=1:
        return lst
    pivot=lst[0]
    bigger=[]
    smaller=[]
    same=[]
    for item in lst:
        if item>pivot:
            bigger.append(item)
        elif item<pivot:
            smaller.append(item)
        else:
            same.append(item)
    return quick_sort(smaller)+same+quick_sort(bigger)
quick_sort([2,3,5,4,6,9,7,9,4,6,6,5,6,4,6])
```

```python
def chebyshev_poly1(n,x):
    if n==0:
        return 1
    elif n==1:
        return x
    else:
        return 2*x*chebyshev_poly1(n-1,x)-chebyshev_poly1(n-2,x)
def my_chebyshev_poly1(n,x):
    # Write your function code here
    y=[]
    for elem in x:
        y.append(chebyshev_poly1(n,elem))
    return y
```

###### Write a function my_ackermann(m,n), where the output is the Ackermann function compuated for m and n.
###### my_ackermann(4,4) is so large that it would be difficult to write down. Although the Ackermann function does not have many practical uses, the inverse Ackermann function has several uses in robotic motion planning.

```python
def my_ackermann(m,n):
    # write your own function code here
    if m==0:
        out=n+1
    elif m>0 and n==0:
        out=my_ackermann(m-1,1)
    elif m>0 and n>0:
        out=my_ackermann(m-1,my_ackermann(m,n-1))
    return out
```

### Problem 4
###### A function, C(n,k), which computes the number of different ways of uniquely choosing k objects from n without repetition, is commonly used in many statistics applications. For example, how many three-flavored ice cream sundaes are there if there are 10 icecream flavors? To solve this problem we would have to compute C(10,3), the number of ways of choosing three unique icecream flavors from 10. The function C is commonly called “n choose k.” You may assume that n and k are integers.
###### If n = k, then clearly C(n,k) = 1 because there is only way to choose n objects from n objects.
###### If k = 1, then C(n,k) = n because choosing each of the n objects is a way of choosing one object from n. For all other cases, C(n,k) = C(n-1,k) + C(n-1,k-1). Can you see why?
###### Write a function my_n_choose_k(n,k) that computes the number of times k objects can be uniquely chosen from n objects without repetition.

```python
def my_n_choose_k(n,k):
    # Write your own function code here
    if n==k:
        out=1
    elif k==1:
        out=n
    else:
        out=my_n_choose_k(n-1,k)+my_n_choose_k(n-1,k-1)
    return out
```

```python
# Output: 10
my_n_choose_k(10,1)
```

```python
# Output: 1
my_n_choose_k(10,10)
```

```python
# Output: 120
my_n_choose_k(10,3)
```

### Problem 5
###### In purchases paid in cash, the seller must return money that was overpaid. This is commonly referred to as “giving change.” The bills and coins required to properly give change can be defined by a recursive relationship. If the amount paid is more than \100 subtracted from the amount paid. If the amount paid is more than \50 subtracted. Similar clauses can be given for every denomination of US currency. The denominations of US currency, in dollars, is 100, 50, 20, 10, 5, 1, 0.25, 0.10, 0.05, and 0.01. For this problem we will ignore the two-dollar bill, which is not in common circulation. You may assume that cost and paid are scalars, and that paid >= cost. The output variable, change, must be a list as shown in the test case.

```python
def my_change(cost, paid):
    # Write your own function code here
    lst=[]
    coins=[100,50,20,10,5,1,0.25,0.10,0.05,0.01]
    for coin in coins:
        while paid-cost-coin>=-1e-5:
            lst.append(coin)
            paid=paid-coin
    return lst
```

```python
# Output: [50.0, 20.0, 1.0, 1.0, 0.25, 0.10, 0.05, 0.01, 0.01, 0.01]
my_change(27.57, 100)
```

### Problem 7
###### The greatest common divisor of two integers a and b is the largest integer that divides both numbers without remainder, and the function to compute it is denoted by gcd(a,b). The gcd function can be written recursively. If b equals 0, then a is the greatest common divisor. Otherwise, gcd(a,b) = gcd(b,a%b) where a%b is the remainder of a divided by b. Assume that a and b are integers.

```python
def my_gcd(a, b):
    # Write your own function code here
    if b==0:
        return a
    return my_gcd(b,a%b)
```

```python
# Output: 2
my_gcd(10, 4)
```

```python
# Output: 11
my_gcd(33, 121)
```

```python
# Output: 1
my_gcd(18, 1)
```

### Problem 8
###### Pascal’s triangle is an arrangement of numbers such that each row is equivalent to the coefficients of the binomial expansion of , where p is some positive integer more than or equal to 1. For example, so the third row of Pascal’s triangle is 1 2 1. Write a function with my_pascal_row(m) where output variable row is the m-th row of the Pascal triangle. You may assume that m is a strictly positive integer.

```python
import numpy as np
def my_pascal_row(m):
    # Write your own function code here
    row=np.zeros((m,m))
    for i in range(0,m):
        row[i,0]=1
        row[i,i]=1
        for j in range(1,i):
            row[i,j]=row[i-1,j-1]+row[i-1,j]
    return row[m-1,:].tolist()
```

```python
# Output: [1]
my_pascal_row(1)
```

```python
# Output: [1, 1]
my_pascal_row(2)
```

```python
# Output: [1, 2, 1]
my_pascal_row(3)
```

```python
# Output: [1, 3, 3, 1]
my_pascal_row(4)
```

```python
# Output: [1, 4, 6, 4, 1]
my_pascal_row(5)
```
